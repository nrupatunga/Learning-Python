```python
#Class and Object Variables
#We have already discussed the functionality part of classes and objects

class Robot:
    """ Represents a robot, with a name"""
    population = 0
    def __init__(self, name):
        """Initializes the data."""
        self.name = name
        print "Initializing {0}".format(self.name)
        Robot.population += 1
        
    def die(self):
        """Robot died"""
        print "{} is being destroyed".format(self.name)
        Robot.population -= 1
        
        if Robot.population == 0:
            print '{} was the last one'.format(self.name)
        else:
            print 'There are still {:d} robots working'.format(Robot.population)
    def say_hi(self):
        """Greeting by the robot
        
        Yeah, they can do that"""
        print 'Greetings, my masters call me {}'.format(self.name)
        
    @classmethod
    def how_many(cls):
        """Prints the current population"""
        print 'We have {} robots'.format(cls.population)
        
droid1 = Robot('R2-D2')
droid1.say_hi()
Robot.how_many()

droid2 = Robot('R3-D3')
droid2.say_hi()
Robot.how_many()
        
```

```python
#Inheritance and Polymorphism
#One of the major benefits of object oriented programming is reuse of code and one of
#the ways this is achieved is through the inheritance mechanism. Inheritance can be
#best imagined as implementing a type and subtype relationship between classes

class SchoolMember:
    '''Represents any school memeber'''
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print 'Initialized schoolmember {}'.format(self.name)
        
    def tell(self):
        '''Tell my details.'''
        print 'Name: "{}" Age: "{}"'.format(self.name, self.age),
        
#specifiy the base class name in a tuple following the the class name, if more than one class is listed
# in the inheritance tuple, then it is called multiple inheritance
class Teacher(SchoolMember):
    '''Represents a teacher.'''
    def __init__(self, name, age, salary):
        #Python does not automatically call the constructor of the base class, we need to call it explicitly
        SchoolMember.__init__(self, name, age)
        self.salary = salary
        print 'Initialized Teacher: {}'.format(self.name)
    def tell(self):
        SchoolMember.tell(self)
        print 'Salary: "{:d}"'.format(self.salary)
        
class Student(SchoolMember):
    '''Represents a student.'''
    def __init__(self, name, age, marks):
        SchoolMember.__init__(self, name, age)
        self.marks = marks
        print 'Initialized student: {}'.format(self.name)
    def tell(self):
        SchoolMember.tell(self)
        print 'Marks: "{:d}"'.format(self.marks)
        
t = Teacher('Mrs. Shrividya', 40, 30000)
print t
s = Student('Nrupatunga', 25, 75)
print s

members = [t, s]
for member in members:
    member.tell()
```
